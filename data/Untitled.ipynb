```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
eps = np.loadtxt("eps.dat")
```

```python
xx = np.array( np.split( eps.T[1], 384 ) )[0]
```

```python
eps2d = np.array( np.split( eps.T[2], 384 ) )
```

```python
plt.plot( eps2d[0], label="y=0" )
plt.plot( eps2d.T[0], label="x=0" )

plt.plot( xx, 2.4/xx**2.0 )
plt.legend()

plt.yscale("log")
plt.xscale("log")

plt.savefig("../figs/eps_corr.pdf", bbox_inches="tight")
```
